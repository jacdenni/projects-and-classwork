```python
%matplotlib inline
from matplotlib import rcParams, cm
from mpl_toolkits.mplot3d import Axes3D
rcParams['savefig.dpi'] = 90
import numpy as np
import matplotlib.pyplot as plt

m1 = 1.
m2 = 5.
s1 = 1.
s2 = 0.2
r = 0.9

def Likelihood(p1, p2):
    A = 1./(2. * s1 * s2 * np.pi * np.sqrt(1 - r**2))
    B = -1./(2 * (1 - r**2)) * ((p1 - m1)**2/s1**2 + (p2 - m2)**2/s2**2 - 2 * r * (p1 - m1) * 
                               (p2 - m2)/(s1 * s2))
    return A * np.exp(B)

fig = plt.figure()
ax = fig.gca(projection = '3d')
X = np.linspace(m1 - 4*s1, m1 + 4*s1)
Y = np.linspace(m2 - 4*s2, m2 + 4*s2)
X, Y = np.meshgrid(X, Y)
Z = Likelihood(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
surf = ax.plot_surface(X, Y, Z, cmap = cm.rainbow, rstride = 1, cstride = 1, linewidth = 0)
plt.show()
```
