```python
%matplotlib inline
import numpy as np
from scipy import interpolate as interp
from scipy.misc import derivative
import matplotlib.pyplot as plt

def onesided_deriv(f, x, h):
    return (f(x + h) - f(x)) / h

def symmetric_deriv(f, x, h):
    return (f(x + h) - f(x - h)) / (2. * h)

def Richardson(f, x):
    h = 0.1 * x
    hList = [h]
    error = 1.
    RRowOld = [symmetric_deriv(f, x, h)]
    flag = 0
    counter = 0
    n = 0
    OptimumRs = [RRowOld[0]]
    ErrorList = []
    while (flag == 0):
        n += 1
        h = h / 2.
        hList.append(h)
        
        RRowNew = []
        Rnew = deriv(f, x, h)
        RRowNew.append(Rnew)
        for m in range(1, n + 1):
            Rnew = RRowNew[m - 1] + (RRowNew[m - 1] - RRowOld[m - 1]) / (
            ((n + 1.) / n)**(2. * (m + 1.)) - 1.) 
            RRowNew.append(Rnew)
        
        error = np.abs(RRowNew[-1] - RRowOld[-1])
        ErrorList.append(error)
        counter += 1
        RRowOld = RRowNew
        OptimumRs.append(RRowNew[-1]) #Keeping track of all the best values for each h
        
        if error <= 1e-15: #Stop of we're hitting machine accuracy with error
            flag = 1
        elif h <= 1e-15: #Stop if we're hitting machine accuracy with h
            flag = 1
        elif counter >= 500: #Preventing an infinite loop
            flag = 1
        else:
            flag = 0
    
    return np.asarray(OptimumRs), np.asarray(hList), np.asarray(ErrorList)
```

```python
def main():
    logk_data, logPk_data = np.loadtxt("pk2.dat", unpack=True)
    kData = 10.**logk_data
    PkData = 10.**logPk_data

    Pkspline = interp.UnivariateSpline(kData, PkData, s = 0, ext = 0)
    PksplinePrime = Pkspline.derivative()
    
    richardson_result = Richardson(Pkspline, 0.52)
    symmetric_result = symmetric_deriv(Pkspline, 0.52, richardson_result[1][1:])
    onesided_result = onesided_deriv(Pkspline, 0.52, richardson_result[1][1:])
    
    symmetric_error = np.fabs(symmetric_result - PksplinePrime(0.52))
    onesided_error = np.fabs(onesided_result - PksplinePrime(0.52))
    
    plt.title('Log error vs. log h for k = 0.52')
    plt.xlabel('log h')
    plt.ylabel('log error')
    plt.plot(np.log10(richardson_result[1][1:]),np.log10(richardson_result[2]), 'r-', 
             label = 'Richardson')
    plt.plot(np.log10(richardson_result[1][1:]), np.log10(symmetric_error), 'b-', 
            label = 'Symmetric')
    plt.plot(np.log10(richardson_result[1][1:]), np.log10(onesided_error), 'g-', 
            label = 'One Sided')
    plt.legend(loc = 0)
    
if __name__ == '__main__':
    main()
```
