```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate as interp
from scipy.misc import derivative


def deriv(f, x, h):
    return (f(x + h) - f(x - h)) / (2. * h)

def Richardson(f, x):
    fprimes = []
    h = 0.1 * x
    hList = [h]
    fprimes.append(deriv(f, x, h))
    err = 1.
    counter = 0
    while (err > 1.0e-8) & (counter <= 100):
        h /= 2.
        hList.append(h)
        fprimes.append(deriv(f, x, h))
        err = np.abs(fprimes[-1] - fprimes[-2])
        counter += 1
        
    interpedFPrimes = interp.UnivariateSpline(np.flipud(hList), np.flipud(fprimes), 
                                              s = 0, ext = 0)
    
    plt.xlabel('log(h)')
    plt.ylabel('P\'({})'.format(x))
    plt.plot(np.log10(np.linspace(1e-11, 0.1)), interpedFPrimes(np.linspace(0., 0.1)), 
            label = 'Richardson')
    
    return [interpedFPrimes(0), err]
```

```python
def main():
    logk_data, logPk_data = np.loadtxt("pk2.dat", unpack=True)
    kData = 10.**logk_data
    PkData = 10.**logPk_data

    Pkspline = interp.UnivariateSpline(kData, PkData, s = 0, ext = 0)
    PksplinePrime = Pkspline.derivative()
    
    plt.figure(0)
    results, error = Richardson(Pkspline, 0.52)
    print 'P\'(0.52) = %E' %PksplinePrime(0.52)
    print 'Richardson P\'(0.52) = %E ± %E' %(results, error)
    
    hPoints = 10**np.arange(-8., 0., 1.)
    
    plt.scatter(np.log10(hPoints), deriv(Pkspline, 0.52, hPoints), color = 'r', 
            label = 'Rough approximation of P\'(0.52)')
    plt.legend(loc = 0)
    plt.title('Approximations of P\'(0.52)')
    
    
    plt.figure(1)
    results, error = Richardson(Pkspline, -0.47)
    print 'P\'(-0.47) = %E' %PksplinePrime(-0.47)
    print 'Richardson P\'(-0.47) = %E ± %E' %(results, error)
    
    hPoints = 10**np.arange(-8., 0., 1.)
    
    plt.scatter(np.log10(hPoints), deriv(Pkspline, -0.47, hPoints), color = 'r', 
            label = 'Rough approximation of P\'(-0.47)')
    plt.legend(loc = 0)
    plt.title('Approximations of P\'(-0.47)')

    
if __name__ == '__main__':
    main()
```
