Create a mock data set corresponding to the sum of two Gaussians, with $\mu_1 = -1.5$, $\mu_2 = -0.5$, $\sigma_1 = \sigma_2 = 0.3$, $p_1 = 0.7$, $p_2 = 0.3$.  

After generating the array of $x$ values, add to each $x$ a random error according to a normal distribution with standard deviation of 0.05.  

Run Gaussian Mixture for this sample with different number of modes $K$ and determine the best value of $K$ according to the AIC and BIC criteria.  

Plot the best-fitting model along with the histogram of the generated sample.  Use adaptive histogram.

Compare the recovered values of the parameters with the original ones and comment on the accuracy of the method.

Generate another sample in the same way, but with the standard deviation of the added error of 0.1.  Comment on the difference in the recovered parameters.

<i>Practical suggestions:</i> Draw a sample of $N=200$ random variables $x$ with the above distribution, using the transformation method.  In this method you draw a random number $r$, uniformly distributed between 0 and 1, and interpolate the cumulative distribution function $F(<x) = \int_{-\infty}^{x} \, p(x) dx$ to find $x$ corresponding to $F(<x) = r$. The cumulative function $F(<x)$ can be expressed as a sum of the complementary error functions.  Be careful with the normalization and the argument of <tt>scipy.special.erf</tt>.  If in doubt, check your result with direct integration of $p(x)$.  

```python
%matplotlib inline
from matplotlib import rcParams
rcParams['savefig.dpi'] = 90
import matplotlib.pyplot as plt
import numpy as np
from scipy import special, stats, interpolate, optimize
from sklearn.mixture import GaussianMixture
from astroML.plotting import hist

np.random.seed(0)

m1 = -1.5; m2 = -0.5; s1 = 0.3; s2 = 0.3; p1 = 0.7; p2 = 0.3;

def fg(x, mu, sig):
    return 1/(sig * np.sqrt(2. * np.pi)) * np.exp(-(x - mu)**2/(2 * sig**2))

def kde_gauss(xRange, data, h):
    N = len(data)
    result = []
    for x in xRange:
        mySum = 0.
        for xi in data:
            mySum += fg(np.fabs((x - xi)/h), 0., 1.)
        result.append(mySum)
    return 1. / (N * h) * np.asarray(result)

def erf_gauss(x, mu, sig):
    return 0.5 * (1. + special.erf((x - mu) / (sig * np.sqrt(2))))

def Phi(x):
    return p1 * erf_gauss(x, m1, s1) + p2 * erf_gauss(x, m2, s2)

def normalize(array):
    A = np.sqrt(sum(array**2))
    return array / A
```

```python
def GetRandomDataWithError(m1, s1, p1, m2, s2, p2, pdf, sig_error, N, plot_title):
    #Generate original sample data
    xRange = np.linspace((m1 - 4*s1), (m2 + 4*s2), 1000)
    Px = pdf(xRange)
    p_spline = interpolate.UnivariateSpline(Px, xRange, s = 0)
    p_sample = np.random.uniform(0, 1, N)
    x_sample = p_spline(p_sample)

    #'Fuzz' the data by adding error
    errors = np.random.normal(0., sig_error, N)
    dataSample = x_sample + errors
    
    X = np.linspace(np.amin(dataSample), np.amax(dataSample), N)
    plt.fill(X, kde_gauss(X, dataSample, .1), color = 'orangered', alpha = 0.7)
    plt.plot(X, p1 * fg(X, m1, s2) + p2 * fg(X, m2, s2), color = 'k', label = 'Model')
    plt.title(plot_title)
    plt.show()
    
    return dataSample


dataSample1 = GetRandomDataWithError(m1, s1, p1, m2, s2, p2, Phi, 0.05, 200, 
                                     '$Data$ $Sample$ $1:$ $\sigma_{error} = 0.05$')
dataSample2 = GetRandomDataWithError(m1, s1, p1, m2, s2, p2, Phi, 0.1, 200, 
                                     '$Data$ $Sample$ $2:$ $\sigma_{error} = 0.1$')
```

```python
def GetGaussianMixture(dataSample): #Generate Gaussian Mixture Models and determine the best K
    K = np.arange(1, 10)
    models = [None for i in K]
    
    X = dataSample.reshape(-1, 1)
    
    for i in range(len(K)):
        models[i] = GaussianMixture(K[i], random_state=1, covariance_type='full', n_init=10).fit(X)
    
    AIC = [m.aic(X) for m in models]
    BIC = [m.bic(X) for m in models]
    logL = [m.score(X) for m in models]
    for i in range(len(K)):
        print 'K = %d  AIC = %.1f  BIC = %.1f  logL = %.3f'%(K[i], AIC[i], BIC[i], logL[i])
    
    gmm_best = models[np.argmin(AIC)]
    
    print 'mu =', gmm_best.means_.flatten()
    print 'sig =', np.sqrt(gmm_best.covariances_.flatten())
    print 'pk =', gmm_best.weights_.flatten()
    if len(gmm_best.means_.flatten()) == 2:
        print 'Errors in mu:', np.fabs(np.array([m1, m2]) - np.sort(gmm_best.means_.flatten()))
        print 'Errors in sig:', np.fabs(np.array([s1, s2]) - gmm_best.covariances_.flatten())
        print 'Errors in pk:', np.fabs(np.array([p1, p2]) - np.sort(gmm_best.weights_.flatten())[::-1])
    else:
        print 'Method did not accurately recover the distribution of the original data sample.'
    
    xmin = np.amin(dataSample)
    xmax = np.amax(dataSample)

    xx = np.linspace(xmin, xmax, num=100).reshape(-1,1)
    logprob = gmm_best.score_samples(xx)
    
    plt.plot(xx, np.exp(logprob), '-k')
    for mu, C, w in zip(gmm_best.means_, gmm_best.covariances_, gmm_best.weights_):
        plt.plot(xx, w * stats.norm.pdf(xx, mu, np.sqrt(C)), '--r')
    xRange = np.linspace(np.amin(dataSample), np.amax(dataSample), 1000)
    plt.fill(xRange, kde_gauss(xRange, dataSample, .1), alpha=0.7, color = '#079E29')
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.show()

print 'DATA SAMPLE 1: _error = 0.05'
GetGaussianMixture(dataSample1)
print 'DATA SAMPLE 2: _error = 0.1'
GetGaussianMixture(dataSample2)
```

The first round of GMM, with sig_error = 0.05, gave pretty accurate results, all within roughly 0.07 for the means and weights. (The standard deviations, however, were generally less accurate, within roughly 0.3 of the target values.) Generally, the second round of GMM with sig_error = 0.1 actually tended to fare slightly better, although not by much. Interestingly, every now and then GMM failed to find the actual underlying distribution, instead producing combinations of 3 or 4 Gaussians. This was probably due to random spikes in the original data collection, which resolved themselves with repeated running of the data collection and/or with larger amounts of data.
