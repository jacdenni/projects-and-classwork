```python
%matplotlib inline
from matplotlib import rcParams
rcParams['savefig.dpi'] = 80
import matplotlib.pyplot as plt
import numpy as np
from astroML.plotting import hist

def tophat(u):
    if u <= 1.:
        return 0.5
    else:
        return 0.
    
def fg(u):
    return 1/np.sqrt(2. * np.pi) * np.exp(-u**2 / 2)
    
def KDE(xRange, X, h, func):
    N = len(X)
    result = []
    for x in xRange:
        mySum = 0.
        for xi in X:
            mySum += func(np.fabs(x - xi) / h)
        result.append(mySum)
    return 1. / (N * h) * np.asarray(result)
    
```

```python
feh = np.loadtxt('gc.dat', unpack = True)[0]
binSize = 0.05
for i in range(0, 5):
    numBins = (1. - (-3.)) / (binSize * 2**i)
    plt.figure(i)
    hist(feh, numBins, (-3., 1.), color = '#183C6D', histtype = 'stepfilled')
    plt.xlabel('[Fe/H]')
    plt.ylabel('Frequency')
    plt.title('Bin size = ' + str(binSize * 2**i))
    plt.show()
```

Initially there appear to be two peaks in the distribution - one around -1.5, and a smaller one around -0.6. As bin size increases, however, the second peak vanishes, and we get a unimodal distribution, peaked a little over -1.5 and skewed right.

```python
xRange = np.linspace(-3, 1, 1000)
kerneledTophat = KDE(xRange, feh, .2, tophat)
kerneledGauss = KDE(xRange, feh, .2, fg)

plt.figure(5)
plt.fill(xRange, kerneledTophat, fc = '#261566')
plt.xlabel('[Fe/H]')
plt.ylabel('Frequency')
plt.title('[Fe/H] Histogram with a Tophat Kernel')
plt.show()

plt.figure(6)
plt.fill(xRange, kerneledGauss, fc = '#15663A')
plt.xlabel('[Fe/H]')
plt.ylabel('Frequency')
plt.title('[Fe/H] Histogram with a Gaussian Kernel')
plt.show()
```
