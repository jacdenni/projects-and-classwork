```python
%matplotlib inline
from matplotlib import rcParams
rcParams['savefig.dpi'] = 90
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import emcee
from matplotlib.colors import LogNorm

m1 = 1.
m2 = 5.
s1 = 1.
s2 = 0.2
r = 0.9
r2 = r * r
mu = np.array([m1, m2])

def lnLikelihood(p):
    p1, p2 = p
    twoLogL = -(((p1 - m1)/s1)**2 + ((p2 - m2)/s2)**2. - 2 * r * (p1 - m1) * \
                (p2 - m2)/ (s1 * s2))/(1 - r2) - 2 * np.log(2. * np.pi * s1 * s2/np.sqrt(1 - r2))
    return twoLogL / 2.

def lnPrior(p):
    p1, p2 = p 
    if not ((m1 - 5*s1) < p1 < (m1 + 5*s1)):
        return -np.inf
    if not ((m2 - 5*s2) < p2 < (m2 + 5*s2)):
        return -np.inf
    return 0.0

def lnPosterior(p):
    return lnLikelihood(p) + lnPrior(p)

def GR_indicator(mwx, swx, mwy, swy, nchain):
    """
    Gelman-Rubin convergence indicator.
    Arguments are 1D arrays of sums of two sample values (mwx,mwy) and their squares (swx,swy), 
      nchain is the number of elements in each walker chain.
    """
    mwxc = mwx/(nchain-1.);  mwyc = mwy/(nchain-1.) 
    swxc = swx/(nchain-1.)-np.power(mwxc,2)
    swyc = swy/(nchain-1.)-np.power(mwyc,2)
    # within chain variance
    Wgrx = np.sum(swxc)/nwalkers; Wgry = np.sum(swyc)/nwalkers
    # mean of the means over Nwalkers
    mx = np.sum(mwxc)/nwalkers; my = np.sum(mwyc)/nwalkers
    # between chain variance
    Bgrx = nchain*np.sum(np.power(mwxc-mx,2))/(nwalkers-1.)
    Bgry = nchain*np.sum(np.power(mwyc-my,2))/(nwalkers-1.)
        
    # Gelman-Rubin R factor
    Rgrx = (1 - 1/nchain + Bgrx/Wgrx/nchain)*(nwalkers+1)/nwalkers - (nchain-1)/(nchain*nwalkers)
    Rgry = (1 - 1/nchain + Bgry/Wgry/nchain)*(nwalkers+1)/nwalkers - (nchain-1)/(nchain*nwalkers)

    return np.array([Rgrx, Rgry])

def plot_samples(xh, yh):
    """
    This routine plots the 2D histogram of MCMC samples and their contour levels,
      along with the contours of the target distribution.
    It uses the same user-supplied 'lnLikelihood' function used to run the emcee sampler,
      'mu' is the array of mean target parameters.
    Arguments are flattened 1D arrays of two samples.
    """
    plt.xlabel(r'$p_1$')
    plt.ylabel(r'$p_2$')
    
    # 2D color histogram
    plt.hist2d(xh, yh, bins=100, norm=LogNorm(), normed=1)
    plt.colorbar()
                      
    # target likelihood level contours
    x = np.arange(np.min(xh)-1, np.max(xh)+1, 0.05)
    y = np.arange(np.min(yh)-1, np.max(yh)+1, 0.05)      
    X, Y = np.meshgrid(x,y)  
    Z = [np.exp(lnLikelihood([x1,x2])) for x1,x2 in zip(X,Y)]
    
    # contours enclosing 68.27 and 99% of the probability density
    dlnL2 = np.array([9.21, 2.30])
    lnLmax = lnLikelihood(mu)
    lvls = np.exp(lnLmax-0.5*dlnL2)
    cs = plt.contour(X,Y,Z, linewidths=1.5, colors='black', norm = LogNorm(), levels = lvls)

    plt.show()
    
    return None
    
def kde_tophat(data, x, h):
    y = (abs(x - data[:,None]) <= h).astype(float)
    return y.sum(0)/(2 * h * len(data))

def pdf_gauss(x, mu, s):
    return np.exp(-((x-mu)/s)**2/2)/(np.sqrt(2*np.pi)*s)
```

```python
#Part 1: Running the MCMC without thinning or burning
ndim = 2
nwalkers = 100
nsteps = 10000
np.random.seed(0)

pos = [(np.random.random(ndim) - 0.5) * np.array([10*s1, 10*s2]) + np.array([m1, m2]) \
       for i in xrange(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnPosterior)
pos, prob, state = sampler.run_mcmc(pos, nsteps / nwalkers)

converged = False
nchain = nsteps/nwalkers

GRIndicators_x = []
GRIndicators_y = []
corr_times_x = []
corr_times_y = []

numLoops = 1

while not converged:
    numLoops += 1
    pos, prob, state = sampler.run_mcmc(pos, nsteps / nwalkers, rstate0 = state)
    mwx = np.array([np.sum(sampler.chain[i, :, 0]) for i in range(nwalkers)])
    mwy = np.array([np.sum(sampler.chain[i, :, 1]) for i in range(nwalkers)])
    swx = np.array([np.sum(sampler.chain[i, :, 0]**2) for i in range(nwalkers)])
    swy = np.array([np.sum(sampler.chain[i, :, 1]**2) for i in range(nwalkers)])
    
    nchain += nsteps/nwalkers
    Rx, Ry = GR_indicator(mwx, swx, mwy, swy, nchain)
    if nchain > 1000:
        GRIndicators_x.append(np.fabs(Rx - 1.))
        GRIndicators_y.append(np.fabs(Ry - 1.))
        corr_time = sampler.get_autocorr_time(c = 3)
        corr_times_x.append(corr_time[0])
        corr_times_y.append(corr_time[1])
    
    if (np.fabs(Rx - 1.) < 0.01) and (np.fabs(Ry - 1.) < 0.01):
        print 'Done'
        converged = True
    elif nchain > 50000: #Hard limit to chain length encoded - otherwise, chain becomes too large for kde_tophat function and my computer to handle. 
        print 'Taking to long to converge. Quitting with accuracy of {0:.5f}%.'.format(100 * \
        max(np.fabs(Rx - 1.), np.fabs(Ry - 1.)))
        converged = True
    if (numLoops % 25) == 0:
        print "p1 = {0:.3f} ± {1:.3f}  p2 = {2:.3f} ± {3:.3f} nchain = {4}"\
          .format(np.mean(sampler.flatchain[:,0]), np.std(sampler.flatchain[:,0]), \
           np.mean(sampler.flatchain[:,1]), np.std(sampler.flatchain[:,1]), nchain)

xData = sampler.flatchain[:, 0]
yData = sampler.flatchain[:, 1]
```

```python
stepRange = np.arange(1001, (numLoops - 10) * 100 + 1001, 100)

plt.plot(stepRange, GRIndicators_x, c = 'r', label = 'p1')
plt.plot(stepRange, GRIndicators_y, c = 'b', label = 'p2')
plt.legend(loc = 0)
plt.xlabel('Steps')
plt.ylabel('GR Indicator')
plt.show()


plt.plot(stepRange, corr_times_x, c = 'r', label = 'p1')
plt.plot(stepRange, corr_times_y, c = 'b', label = 'p2')
plt.legend(loc = 0)
plt.xlabel('Steps')
plt.ylabel('Autocorrelation time')
plt.show()

calc_m1 = np.mean(xData)
calc_m2 = np.mean(yData)
calc_s1 = np.std(xData)
calc_s2 = np.std(yData)

print "p1 = {0:.3f} ± {1:.3f}  p2 = {2:.3f} ± {3:.3f} nchain = {4}"\
    .format(calc_m1, calc_s1, calc_m2, calc_s2, nchain)    
print 'Correlation Coefficients:', stats.pearsonr(xData, yData)[0]
```

```python
xRange = np.linspace(np.amin(xData), np.amax(xData))
yRange = np.linspace(np.amin(yData), np.amax(yData))


plt.fill(xRange, kde_tophat(xData, xRange, 0.05), color = '#292A64')
plt.plot(xRange, pdf_gauss(xRange, calc_m1, calc_s1), c = 'r')
plt.xlabel('p1')
plt.show()


plt.fill(yRange, kde_tophat(yData, yRange, 0.05), color = '#292A64')
plt.plot(yRange, pdf_gauss(yRange, calc_m2, calc_s2), c = 'r')
plt.xlabel('p2')
plt.show()

plot_samples(xData, yData)
```

There seems to be very good correspondence between the Gaussian and the kde_tophat function. This may be due to the small tophat size of the KDE.

```python
nthin = int(np.amax(sampler.get_autocorr_time(c = 3)))
nburn = int(nwalkers * nthin)

p1_thinned = []
p2_thinned = []

for i in range(nwalkers):
    p1_thinned.append(sampler.chain[i, nburn::nthin, 0])
    p2_thinned.append(sampler.chain[i, nburn::nthin, 1])

p1_thinned = np.array(p1_thinned).flatten()
p2_thinned = np.array(p2_thinned).flatten()

plot_samples(p1_thinned, p2_thinned)
```

This graph looks much like that of the full chain, except with fewer outlying values. Instead, almost all of the data falls nicely within an ellipse.

```python
print 'Thinned Chain:'
print 'p1 = {0:.3f} ± {1:.3f}  p2 = {2:.3f} ± {3:.3f}'\
    .format(np.mean(p1_thinned), np.std(p1_thinned), \
    np.mean(p2_thinned), np.std(p2_thinned))
print 'Correlation Coefficient:', stats.pearsonr(p1_thinned, p2_thinned)[0]

print '\nFull Chain:'
print "p1 = {0:.3f} ± {1:.3f}  p2 = {2:.3f} ± {3:.3f} nchain = {4}"\
    .format(calc_m1, calc_s1, calc_m2, calc_s2, nchain)    
print 'Correlation Coefficients:', stats.pearsonr(xData, yData)[0]

print '\nTarget Values:'
print "p1 = {0:.3f} ± {1:.3f}  p2 = {2:.3f} ± {3:.3f}"\
    .format(m1, s1, m2, s2)    
print 'Correlation Coefficient:', r
```

Both the thinned chain and full chain values are very similar, with slightly better correlation for the thinned chain. They do, however, agree very well with the target values.
