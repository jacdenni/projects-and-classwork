```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt

def GenerateRandomWalk(t):
    x = 0.0
    for i in np.arange(t):
        randNum = rand.random_integers(0, 1)
        if randNum == 0:
            x += -1.0
        elif randNum == 1:
            x += 1.0
    return x

def GetSetsofWalks(numSets, t):
    X = []
    for n in np.arange(numSets):
        X.append(GenerateRandomWalk(t))
    return np.array(X)

def main():
    meanXs  = []
    meanX2s = []
    tRange = np.arange(0.0, 10.0)
    for t in tRange:
        randomWalks = GetSetsofWalks(500, t)
        meanXs.append(np.mean(randomWalks))
        meanX2s.append(np.mean(randomWalks**2))
    plt.figure(0)
    plt.clf()
    plt.ylim(-1.0, 1.0)
    plt.plot(tRange, meanXs)
    plt.xlabel('t')
    plt.ylabel('<x>')
    plt.savefig('Homework 8 - Average X for a Random Walk.png')
    
    plt.figure(1)
    plt.clf()
    plt.plot(tRange, meanX2s)
    plt.xlabel('t')
    plt.ylabel('<x**2>')
    plt.title('Position of a Random Walk Squared /n (Averaged over 500 Walks)')
    plt.savefig('Homework 8 - Average X**2 for a Random Walk.png')


if __name__ == '__main__':
    main()
```

```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt

def RandomSteps(size):
    steps = []
    for i in range(size):
        randNum = rand.random_integers(0, 1)
        if randNum == 0:
            steps.append(-1.0)
        elif randNum == 1:
            steps.append(1.0)
    return np.array(steps)

def GenerateRandomWalk(size):
    X = 0.0
    Steps = RandomSteps(25)
    for step in Steps:
        X += step
    return X

def main():
    tRange = np.arange(0.0, 15.0, 1.0)
    
    AvgXsforTs = []
    AvgX2sforTs = []
    for t in tRange:
        print t
        Xs = []
        X2s = []
        for i in range(50000):
            tempX = GenerateRandomWalk(t)
            Xs.append(tempX)
            X2s.append(tempX**2)
        AvgXsforTs.append(np.mean(Xs))
        AvgX2sforTs.append(np.mean(X2s))
    plt.clf()
    plt.plot(tRange, AvgXsforTs)
    plt.savefig('Homework 8 - Monte Carlo II 2.png')
    
    plt.clf()
    plt.plot(tRange, AvgX2sforTs)
    plt.savefig('Homework 8 - Monte Carlo II 3.png')

if __name__ == '__main__':
    main()
```
