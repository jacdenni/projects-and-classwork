```python
import numpy as np
import matplotlib.pyplot as plt

def rho(x, y):
    return np.sin(np.pi * x) * np.sin(2.0 * np.pi * y)

def poissonPhi(x, y):
    return -4.0 * np.pi * rho(x, y)

def Relaxation((x0, y0), (xf, yf), dx, dy, poisson, guess, tol):
    Nx = int((xf - x0) / dx)
    Ny = int((yf - y0) / dy)
    Phi = np.empty(((Ny + 1), (Nx + 1)))
    PhiPrevious = guess
    for i in range(Nx + 1):
        Phi[0][i] = i * dx
        PhiPrevious[0][i] = i * dx
        Phi[-1][i] = i * dx
        PhiPrevious[-1][i] = i * dx
    for i in range(Ny + 1):
        Phi[i][-1] = 1.0
        PhiPrevious[i][-1] = 1.0
    
    difference = 1.0
    iterationNum = 0
    while (difference > tol) & (iterationNum <= 10000):
        #if ((iterationNum % 100) == 0):
        #    print iterationNum
        for i in range(1, Ny):
            for j in range(1, Nx):
                Phi[i][j] = 0.25 * (PhiPrevious[i + 1][j] + PhiPrevious[i - 1][j] + PhiPrevious[i][j + 1]
                                    + PhiPrevious[i][j - 1] - poisson(j * dy, i * dx))
        difference = np.max(np.abs(Phi - PhiPrevious))
        Phi, PhiPrevious = PhiPrevious, Phi
        iterationNum += 1
    print 'It took', iterationNum, 'iterations to get a solution.'  
    return Phi

def Interpolation(A, iterationNum):
    xSizeNew = 2**iterationNum + 1
    ySizeNew = 2**iterationNum + 1
    interpolatedA = np.zeros((xSizeNew, ySizeNew))
    for i in range(ySizeNew):
        #print 'Row', i
        for j in range(xSizeNew):
            #print 'Column', j
            if (i % 2 == 0):
                if (j % 2 == 0):
                    interpolatedA[i][j] = A[i/2][j/2]
                elif (j % 2 == 1):
                    interpolatedA[i][j] = 0.5 * (A[i/2][(j - 1)/2] + A[i/2][(j + 1)/2])
            elif (i % 2 == 1):
                if (j % 2 == 0):
                    interpolatedA[i][j] = 0.5 * (A[(i - 1)/2][j/2] + A[(i + 1)/2][j/2])
                elif (j % 2 == 1):
                    interpolatedA[i][j] = 0.25 * (A[(i - 1)/2][(j - 1)/2] + A[(i - 1)/2][(j + 1)/2] + 
                                                  A[(i + 1)/2][(j - 1)/2] + A[(i + 1)/2][(j + 1)/2])
    return interpolatedA


zerosGuess = np.zeros((3, 3))
dx = 0.5
dy = 0.5
iterationNum = 1
phiGuess = Relaxation((0.0, 0.0), (1.0, 1.0), dx, dy, poissonPhi, zerosGuess, 1.0e-5)
dx = 0.25
dy = 0.25
iterationNum = 2

phiGuesses = [phiGuess]

while (dx > 2.**(-7)):
    print 'Iteration Number:', iterationNum
    newPhiGuess = Interpolation(phiGuess, iterationNum)
    print newPhiGuess
    phiGuess = Relaxation((0.0, 0.0), (1.0, 1.0), dx, dy, poissonPhi, newPhiGuess, 1.0e-5)
    phiGuesses.append(phiGuess)
    dx = dx / 2.0
    dy = dy / 2.0
    iterationNum += 1

plt.imshow(phiGuess)
plt.show()
```
