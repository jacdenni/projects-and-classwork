```python
from numpy import *
from math import *
import time
import pylab as pl
climateFile = open('climate.txt', 'r')
climateData = fliplr(rot90(loadtxt(climateFile, usecols = (8, 10), skiprows = 2), k = 3))
data = climateData[1][-1024:]
dateRange1024 = climateData[0][-1024:]
#print(data)
def expon(N,k):
    return e**(2*pi*1j*k/N)

def dft(y):
    N=len(y)
    c=zeros(N, complex)
    for k in range(N):
        for n in range(N):
            c[k]+=y[n]*expon(N, -k*n)
    return c.tolist()

def fftr(y):
    n = len(y)
    
    Feven=[]
    Fodd=[]
    if n == 1:
        return y
    else:
        for i in arange(0,n,2):
            print(type(Feven))
            Feven.append(y[i])
            Feven=fftr(Feven)
        for i in arange(1,n,2):
            Fodd.append(y[i])
            Fodd=fftr(Fodd)
        combined=zeros(n, complex)
        for m in range(n//2):
            combined[m] = Feven[m] + expon(n, -m) * Fodd[m]
            combined[m + n//2] = Feven[m] - expon(n, -m) * Fodd[m]
        return combined
    
print(fftr(data))
pl.subplot(1,1,1)
pl.plot(range(1024), fftr(data), 'b', label='Real part, FFT')
pl.plot(range(1024), dft(data), 'r', label='Real part, DFT')
pl.axis([0,1024, -45000, 30000])
pl.subplot(2,1,2)
for i in range(len(data)):  
    pl.plot(i, fftr(data)[i].imag, 'b', label='Imaginary part, FFT')
    pl.plot(i, dft(data)[i].imag, 'r', label='Imaginary part, DFT')
pl.show()
##dft_time=[]
##fftr_time=[]
##fft_time=[]
##lengthlist=[]
##m=3
##elapsedtimefft=0
##elapsedtimefftr=0
###for m in range(3,11):
##while elapsedtimefft<1:
##    data=random.random(2**m)
####    start=time.clock()
####    dft(data)
####    end=time.clock()
####    dft_time.append((end-start))
####    start=time.clock()
####    fftr(data)
####    end=time.clock()
####    fftr_time.append((end-start))
##    start=time.clock()
##    fft.fft(data)
##    end=time.clock()
##    elapsedtimefft=end-start
##    m+=1
##print('The packaged method can transform', len(data), 'data points in under 1 second')
##m=3
##while elapsedtimefftr<1:
##    data=random.random(2**m)
##    start=time.clock()
##    fftr(data)
##    end=time.clock()
##    elapsedtimefftr=end-start
##    m+=1
##print('The implemented fft method can transform', len(data), 'data points in under 1 second')
##elapsedtimedft=0
##m=3
##while elapsedtimedft<1:
##    data=random.random(2**m)
##    start=time.clock()
##    dft(data)
##    end=time.clock()
##    elapsedtimedft=end-start
##    m+=1
##print('The implemented dft method can transform', len(data), 'data points in under 1 second')
####print('DFT time data in seconds:', dft_time)
####pl.loglog(lengthlist, dft_time, 'bd', label='DFT')
####pl.loglog(lengthlist, fftr_time, 'gd', label='Implemented FFT')
####pl.loglog(lengthlist, fft_time, 'rd', label='Packaged FFT')
####pl.legend(loc='upper left')
####pl.xlabel('log(data size)')
####pl.ylabel('log(time (s))')
####pl.title('Log-Log plot of time taken vs data size')
####pl.grid()
####pl.show()

```
