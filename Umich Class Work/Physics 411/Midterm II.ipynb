```python
import numpy as np
import matplotlib.pyplot as plt

#Constants
gamma = 0.0
g = 9.81

def RDoublePrime(V):
    A = np.array([[-gamma, 0.0], 
                  [0.0, -gamma - g]])
    v = np.linalg.norm(V)
    return np.dot(A, V) * v

def VectorF(RVvector, t):
    rPrime = RVvector[1]
    vPrime = RDoublePrime(RVvector[1])
    return np.array([rPrime, vPrime])

def ForwardEuler(r0, v0, f, a, b, N):
    h = (b - a) / N
    T = np.arange(a, b, h)
    RV = np.zeros((N + 1, 2, 2, 1))
    RV[0] = np.array([r0, v0])
    for i in range(N):
        RV[i + 1] = RV[i] + h * f(RV[i], T[i])
    return RV

def RungeKutta(r0, v0, f, a, b, N):
    h = (b - a) / N
    T = np.arange(a, b, h)
    RV = np.zeros((N + 1, 2, 2, 1))
    RV[0] = np.array([r0, v0])
    for i in range(N):
        k1 = h * f(RV[i], T[i])
        k2 = h * f(RV[i] + .5 * k1, T[i] + 0.5 * h)
        k3 = h * f(RV[i] + .5 * k2, T[i] + 0.5 * h)
        k4 = h * f(RV[i] + k3, T[i] + h)
        RV[i + 1] = RV[i] + 1.0 / 6.0 * (k1 + 2.0 * k2 + 2.0 * k3 + k4)
    return RV

def ExtractXandZ(RVarray):
    xResults = []
    zResults = []
    for rv in RVarray:
        if (rv[0][0][0] != float('Inf')) & (rv[0][1][0] != -float('Inf')):
            xResults.append(rv[0][0][0])
            zResults.append(rv[0][1][0])
    xResults = np.array(xResults)
    zResults = np.array(zResults)
    return np.array([xResults, zResults])


v0 = np.array([[5.0 * np.cos(np.pi/4)],
               [5.0 * np.sin(np.pi/4)]])
r0 = np.array([[0.0],
               [3.0]])

allEulerResults = ForwardEuler(r0, v0, VectorF, 0.0, 1.0, 10)
allRungeKuttaResults = RungeKutta(r0, v0, VectorF, 0.0, 1.0, 10)
motionEulerResults = ExtractXandZ(allEulerResults)
motionRKResults = ExtractXandZ(allRungeKuttaResults)

print motionEulerResults[0]
print motionRKResults[0]

plt.clf()
plt.plot(motionEulerResults[0], motionEulerResults[1])
plt.xlabel('X')
plt.ylabel('Z')
plt.savefig('Midterm 2 - Friction - Euler.png')

plt.clf()
plt.plot(motionRKResults[0], motionRKResults[1])
plt.xlabel('X')
plt.ylabel('Z')
plt.savefig('Midterm 2 - Friction - Runge Kutta.png')
```
