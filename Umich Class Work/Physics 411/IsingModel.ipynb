```python
import numpy as np
import numpy.random as rand

def MakeLattice(L):
    A = np.ones([L, L]) # spins will be either 1 or -1, so makes sense to set initial state to be all ones.
    return A

def Hamiltonian(A): # Use dimensionless units - so set J = 1
    L = A.shape[0]
    E = 0.
    for i in range(L):
        for j in range(L):
            E += -A[i][j] * A[(i + 1) % L][j]
            E += - A[i][j] * A[i][(j + 1) % L] # This allows us to deal with our periodic boundary conditions - 
                                               # if we had gone to the left, we would have had to add +L to everything, 
                                               # then add the %
    return E
    
def step(A, inverse_T):
    L = A.shape[0]
    x = rand.random_integers(0, int(L) - 1)
    y = rand.random_integers(0, int(L) - 1)
    
    dE = EnergyDifference(A, x, y)
    
    if (min(1, np.exp(-dE * inverse_T)) > rand.random()):
        A[x][y] *= -1
    return None
    
def EnergyDifference(A, x, y):
    E_old = Hamiltonian(A)
    A[x][y] *= -1.
    E_new = Hamiltonian(A)
    A[x][y] *= -1. #flipping back so as to not actually change configuration - we'll want to accept or reject this move
                   #based on a Metropolis criterion
    return E_new - E_old

def Magnetization(A):
    L = A.shape[0]
    return np.abs(1.0 * np.sum(A) / (L**2))

def main():
    A = MakeLattice(10)
    T = 5
    
    ntherm = 1000
    '''
    for i in range(ntherm): # if temp is much lower than phase transition state, then nothing much will happen - but the 
                            # more we run this (i.e. the higher the temperature), we'll start to see fractal behaviour - 
                            # the spin config will become completely random
        step(A, 1.0/T)
    '''
    nstep = 10000 # Can also run for 10000 steps, get observables for an energy.
    Eaverage = 0.
    maverage = 0
    for i in range(nstep):
        step(A, 1./T)
        Eaverage += Hamiltonian(A)
        maverage += Magnetization(A)
    
    print 'Average energy is:', Eaverage / nstep
    print 'Average magnetization is:', maverage / nstep
    print 'Temperature:', T
    print A
    
if __name__ == '__main__':
    main()
```
