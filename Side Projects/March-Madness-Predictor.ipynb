```python
import numpy as np

class Team:
    def __init__(self, seed, name):
        assert type(seed * 1.) is float, "seed is not a number: %r" % seed
        assert type(name) is str, "name is not a string: %r" % name
        self.seed = seed
        self.name = name
    
    def __str__(self):
        return self.name

def getOdds(team1, team2): #returns odds that first team will win the matchup
    x = team1.seed - team2.seed
    return (1/(1 + np.exp(0.25 * x))) 

def PlayGame(team1, team2): #simulates playing a game
    odds = getOdds(team1, team2)
    return team1 if np.random.random() < odds else team2

#Returns: new Division
#Effects: plays a round.
def PlayRound(Division):
    N = len(Division)
    NewDivision = []
    for match in range(0, N, 2):
        winner = PlayGame(Division[match], Division[match + 1])
        print("{0} vs. {1}: {2} wins.".format(Division[match], Division[match + 1], winner))
        NewDivision.append(winner)
        
    return NewDivision

def PlayTournament(South, East, West, Midwest):
    N = len(South)
    while (N*4 >= 8):
        if N*4 == 16:
            print('\n --- Sweet Sixteen --- \n')
        elif N*4 == 8:
            print('\n --- Elite Eight --- \n')
        else:
            print('\n --- Round of {} --- \n'.format(N*4))
    
        print('\nSouth Division\n')
        South = PlayRound(South)
        
        print('\nEast Division\n')
        East = PlayRound(East)
        
        print('\nWest Division\n')
        West = PlayRound(West)
        
        print('\nMidwest Division\n')
        Midwest= PlayRound(Midwest)
        
        N = len(South)
    
    #Play Final Four
    print('\n --- Final Four ---\n')
    SouthWest_Champ = PlayRound([South[0], West[0]])[0]
    EastMidwest_Champ = PlayRound([East[0], Midwest[0]])[0]
    
    #Play Championship Match
    print('\n --- Championship Match --- \n')
    Champ = PlayRound([SouthWest_Champ, EastMidwest_Champ])[0]
    
    print('\n', Champ, 'wins the 2018 NCAA Baskbetball Tournament!')
```
