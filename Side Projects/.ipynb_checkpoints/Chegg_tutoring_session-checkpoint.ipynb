```python
%matplotlib inline
from matplotlib import rcParams
rcParams['savefig.dpi'] = 90
import numpy as np
from scipy import interpolate as interp
import matplotlib.pyplot as plt

np.random.seed(0)

X = np.arange(0, 10)
Y = np.random.sample((10,))
f = interp.interp1d(X, Y, 'cubic')

X2 = np.linspace(0, 9, 100)
plt.plot(X2, f(X2))
plt.scatter(X, Y)
plt.show()
```
