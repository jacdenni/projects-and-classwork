```python
%matplotlib inline
from matplotlib import rcParams
rcParams['savefig.dpi'] = 90
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 4*np.pi, num = 400)
accumulated_function = np.sin(X)
plt.plot(X, accumulated_function)
plt.title('k = 1')
plt.show()
for k in np.arange(3, 51, 2):
    accumulated_function += 1/k * np.sin(k * X)
    plt.plot(X, 1/k * np.sin(k * X))
    plt.plot(X, accumulated_function)
    plt.title("k = {}".format(k))
    plt.show()
```
